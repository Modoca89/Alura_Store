### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt
import folium
from folium.plugins import MarkerCluster

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.info()
```

#1. Análisis de facturación



```python
tienda_1_total = tienda['Precio'].sum()
tienda_2_total = tienda2['Precio'].sum()
tienda_3_total = tienda3['Precio'].sum()
tienda_4_total = tienda4['Precio'].sum()
```

```python
print(f'El precio de la tienda 1 es : ${(tienda_1_total):,.2f} ')
print(f'El precio de la tienda 2 es : ${(tienda_2_total):,.2f} ')
print(f'El precio de la tienda 3 es : ${(tienda_3_total):,.2f} ')
print(f'El precio de la tienda 4 es : ${(tienda_4_total):,.2f} ')
```

# 2. Ventas por categoría

#Visualización de gráficos

## Gráfico de Ingresos por tienda

```python
# Datos
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
totales = [tienda_1_total, tienda_2_total, tienda_3_total, tienda_4_total]

# Crear gráfico
plt.figure(figsize=(10, 6))
bars = plt.bar(tiendas, totales, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'])
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height,
             f'${height:,.2f}',  # Formato: $1,234.56
             ha='center', va='bottom', fontsize=10)
plt.title('Ventas Totales por Tienda', fontsize=14, pad=20)
plt.xlabel('Tiendas', fontsize=12)
plt.ylabel('Total en Pesos ($)', fontsize=12)
plt.xticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.6)
min_y = min(totales) * 0.95
max_y = max(totales) * 1.05
plt.ylim(min_y, max_y)
plt.show()
```

## Gráfica de promedios

```python
# Datos
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
promedios = [3.98, 4.04, 4.05, 4.0]  # Reemplaza con tus variables reales

# Configuración del gráfico
plt.figure(figsize=(10, 5))
plt.plot(tiendas, promedios, marker='o', linestyle='--', color='#FF6B6B', linewidth=2, markersize=10, label='Calificación promedio')

# Añadir etiquetas con valores exactos (3 decimales)
for i, prom in enumerate(promedios):
    plt.text(i, prom + 0.005, f'{prom:.3f}', ha='center', va='bottom', fontsize=10, bbox=dict(facecolor='white', alpha=0.8, edgecolor='none'))

# Personalización avanzada
plt.title('Diferencias en Calificaciones Promedio por Tienda', fontsize=14, pad=20)
plt.xlabel('Tiendas', fontsize=12)
plt.ylabel('Calificación (escala 1-5)', fontsize=12)
plt.ylim(3.96, 4.08)  # ¡Zoom para resaltar diferencias!
plt.grid(axis='y', linestyle=':', alpha=0.6)

# Línea horizontal de referencia (opcional)
plt.axhline(y=4.0, color='gray', linestyle=':', label='Referencia (4.0)')
plt.legend()

plt.show()
```

## Análisis de desempeño geográfico

```python
todas_tiendas = pd.concat([tienda, tienda2, tienda3, tienda4])
mapa = folium.Map(location=[4.5709, -74.2973], zoom_start=6)

# Añadimos capas de marcadores agrupados para cada tienda
cluster1 = MarkerCluster(name="Tienda 1").add_to(mapa)
cluster2 = MarkerCluster(name="Tienda 2").add_to(mapa)
cluster3 = MarkerCluster(name="Tienda 3").add_to(mapa)
cluster4 = MarkerCluster(name="Tienda 4").add_to(mapa)

# Función para añadir marcadores con popup informativo
def añadir_marcadores(datos, cluster, color, tienda_nombre):
    for idx, row in datos.iterrows():
        # Información para mostrar en el popup
        popup_text = f"""
        <b>Tienda:</b> {tienda_nombre}<br>
        <b>Producto:</b> {row['Producto']}<br>
        <b>Categoría:</b> {row['Categoría del Producto']}<br>
        <b>Precio:</b> ${row['Precio']:,.0f}<br>
        <b>Calificación:</b> {row['Calificación']}/5
        """

        folium.CircleMarker(
            location=[row['lat'], row['lon']],
            radius=5,
            color=color,
            fill=True,
            fill_color=color,
            fill_opacity=0.7,
            popup=folium.Popup(popup_text, max_width=300)
        ).add_to(cluster)

# Añadimos los datos de cada tienda (con colores distintos)
añadir_marcadores(tienda, cluster1, 'blue', 'Tienda 1')
añadir_marcadores(tienda2, cluster2, 'green', 'Tienda 2')
añadir_marcadores(tienda3, cluster3, 'red', 'Tienda 3')
añadir_marcadores(tienda4, cluster4, 'purple', 'Tienda 4')

# Añadimos control de capas para mostrar/ocultar tiendas
folium.LayerControl().add_to(mapa)

# Añadimos título al mapa
title_html = '''
    <h3 align="center" style="font-size:16px">
    <b>Distribución geográfica de ventas por tienda</b>
    </h3>
'''
mapa.get_root().html.add_child(folium.Element(title_html))

# Mostramos el mapa
mapa
```
